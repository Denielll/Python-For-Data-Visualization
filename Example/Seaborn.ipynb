```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.distplot(tips["total_bill"],kde = False,bins = 20)
```

```python
sns.jointplot(x ="total_bill", y  ="tip", data = tips,kind="reg")  # 好猛喔！
```

```python
sns.pairplot(tips,hue = "sex",palette="coolwarm")
```

```python
sns.rugplot(tips["total_bill"])
```

```python
sns.kdeplot(tips["total_bill"])
```

# Category Data

```python
import numpy as np
```

```python
sns.barplot(x="sex",y="total_bill",data = tips,estimator=np.std)
```

```python
sns.countplot(x="sex",data=tips)
```

```python
sns.boxplot(x="day",y="total_bill",data=tips,hue="sex")   # 類別資料的分布比較
```

```python
sns.violinplot(x="day",y="total_bill",data = tips,hue="sex",split=True)  #同 boxplot功能
```

```python
sns.stripplot(x="day",y="total_bill",data = tips,jitter=True,hue = "sex")  # 同 boxplot功能
```

```python
sns.swarmplot(x="day",y="total_bill",data = tips,color="black")  # 同boxplot功能，但不好用
sns.violinplot(x="day",y= "total_bill",data=tips,color="red")
```

```python
sns.factorplot(x="day",y= "total_bill",data= tips,kind = "swarm") # 萬能公式，改變變數 kind可以改變圖形
```

# Pivot Table  

```python
flights = sns.load_dataset("flights")
```

```python
flights.head()
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc,annot= True,cmap="coolwarm")
```

```python
g = flights.pivot_table(index="month",columns="year",values = "passengers")  # 好猛！ 樞紐分析功能把month從column轉到index
```

```python
sns.heatmap(g,cmap="magma",linewidths=0.3,linecolor="white")
```

```python
sns.clustermap(g,standard_scale=1)
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

# 有兩種方法建造matplotlib的圖： 1.函數法 2.物件導向法

# 函數法

```python

# plt.show() # 如果不在jupyter notebook執行 matplotlib的圖的話，要打這一行指令， e.g.終端機



plt.xlabel("x")
plt.ylabel("y")
plt.title("hshsh")
plt.plot(x,y,"r")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

# 物件導向法  有點難，有點特殊   ( 三步驟，1叫fig  2叫axes也就是軸 3用軸.plot ）

```python
fig = plt.figure() # 把fig 想像成一格空白的 canvas


a = fig.add_axes([0.1,0.1,0.8,0.8]) # 左到右位置，下到上位置，長度，高度
b = fig.add_axes([0.1,0.2,0.3,0.3])

a.plot(x,y)
a.set_title("Large")

b.plot(y,x)
b.set_title("smaller")





# 記住，如果要用這種方法弄出圖來，那個cell裡的第一行一定要包含 fig = plt.figure()，把它想像成空白 canvas

```

# 繼續物件導向... Subplots 和 axes觀念

```python
fig,axes = plt.subplots(nrows=1,ncols=2)   # 和上面不同的是，這行直接叫出 axes （ 一個物件），等同上面的a或b

axes[0].plot(x,y)  # axes是 array！

plt.tight_layout() # 讓兩張圖不會overlapping
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label="X Square")
ax.plot(x,x**3,label="X Cube")

ax.legend(loc = 0)



 # 兩步驟 為 圖標上label
```

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
sns.distplot(tips["total_bill"],kde = False,bins = 20)
```

```python
sns.jointplot(x ="total_bill", y  ="tip", data = tips,kind="reg")  # 
```

```python
sns.pairplot(tips,hue = "sex",palette="coolwarm")
```

```python
sns.rugplot(tips["total_bill"])
```

# Category Data

```python
import numpy as np
```

```python
sns.barplot(x="sex",y="total_bill",data = tips,estimator=np.std)
```

```python
sns.countplot(x="sex",data=tips)
```

```python
sns.boxplot(x="day",y="total_bill",data=tips,hue="sex")   # 
```

```python
sns.violinplot(x="day",y="total_bill",data = tips,hue="sex",split=True)  # boxplot
```

```python
sns.stripplot(x="day",y="total_bill",data = tips,jitter=True,hue = "sex")  #  boxplot
```

```python
sns.swarmplot(x="day",y="total_bill",data = tips,color="black")  # boxplot
sns.violinplot(x="day",y= "total_bill",data=tips,color="red")
```

```python
sns.factorplot(x="day",y= "total_bill",data= tips,kind = "swarm") #  kind
```

# Pivot Table  

```python
flights = sns.load_dataset("flights")
```

```python
flights.head()
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc,annot= True,cmap="coolwarm")
```

```python
g = flights.pivot_table(index="month",columns="year",values = "passengers")  #  monthcolumnindex
```

```python
sns.heatmap(g,cmap="magma",linewidths=0.3,linecolor="white")
```

```python
sns.clustermap(g,standard_scale=1)
```
